```python
import pandas as pd
import numpy as np
```

```python
#nomor 1

df1 = pd.DataFrame({
    "name": ["Anna", "Dane", "David", "Kevin", "Joe", "Rose"],
    "weight": [45.2, 50.5, 62.1, 64.0, 59.0, np.nan],
    "height": [155, 160, 162, 170, 167, 159],
    "birth_year": [2000,1997,1996,1987,1998,2003],
    "gender": ["female", "male", "male", "male", "male", "female"],
    "city": ["New York", "Chicago", "Austin", "New York", np.nan, "New York"]
    })
df1
```

```python
#nomor 2

df2 = pd.DataFrame({
    "nick_name": ["Anna", "David", "Dane", "Rose"],
    "full_name": ["Anna Levis", "David John", "Dane Sebastian", "Rose Kimberly"]
    })
df2
```

```python
df3 = df1.merge(df2, how = "left", left_on = "name", right_on = "nick_name")
df3
```

```python
#nomor 3

df3 = df1.merge(df2, how = "left", left_on = "name", right_on = "nick_name").drop(columns = "nick_name")
df3
```

```python
df3["full_name"] = df3["full_name"].fillna(df3["full_name"].mode()[0])
df3
```

```python
#nomor 4

def bmi (weight, height):
    bmi = weight/(height/100)**2
    return bmi

df4= pd.DataFrame({ 
            "bmi" : [bmi(45.2,155), bmi(50.5,160), bmi(62.1,162), bmi(64.0,170), bmi(59.0,167), bmi(np.nan,159)]
})
df4
```

```python
df5 = pd.concat([df3, df4], axis = 1)
df5
```

```python
#nomor 5

def bmi_class (bmi, kelas):
    if kelas == True:
        return ("NaN")
    else:
        if (bmi < 18.5):
            print("underweight") 
        elif ( bmi >= 18.5 and bmi < 25):
            print("normal weight")
        elif ( bmi >= 25 and bmi < 30):
            print("overweight")
        elif ( bmi >=30 and bmi < 35):
            print("obesity class 1")
        elif ( bmi >=35 and bmi < 40):
            print("obesity class 2")
        elif ( bmi >= 40):
            print("obesity class 3")
        elif ( bmi <= np.nan ):
            print ("NaN")
```

```python
bmi_class(18.81,False)
```

```python
df6 = pd.DataFrame({ 
        "bmi_class" : ["Normal weight", "Normal weight", "Normal weight", "Normal weight", "Normal weight", np.nan]
})
df6
```

```python
df7 = pd.concat([df5, df6], axis = 1)
df7
```
